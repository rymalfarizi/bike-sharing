### Cleaning Data

```python
# Mengubah nilai normalisasi menjadi nilai real
day_bike_df['temp'] = day_bike_df['temp']*41
day_bike_df['atemp'] = day_bike_df['atemp']*50
day_bike_df['hum'] = day_bike_df['hum']*100
day_bike_df['windspeed'] = day_bike_df['windspeed']*67

hour_bike_df['temp'] = hour_bike_df['temp']*41
hour_bike_df['atemp'] = hour_bike_df['atemp']*50
hour_bike_df['hum'] = hour_bike_df['hum']*100
hour_bike_df['windspeed'] = hour_bike_df['windspeed']*67
```

```python
# Mengubah tipe data object menjadi datetime
day_bike_df['dteday'] = pd.to_datetime(day_bike_df['dteday'])
hour_bike_df['dteday'] = pd.to_datetime(hour_bike_df['dteday'])
```

## Exploratory Data Analysis (EDA)

### Explore ...

Waktu apa yang paling banyak pesepeda?

```python
# Mengubah domain waktu menjadi per bulan
monthly_bike_df = hour_bike_df.resample(rule='M', on='dteday').cnt.sum()
monthly_bike_df.index = monthly_bike_df.index.strftime('%Y-%m')
monthly_bike_df = monthly_bike_df.reset_index()
monthly_bike_df.columns = ['dtemonth','cnt']
monthly_bike_df
```

Cuaca apa yang cocok bagi para pesepeda?

```python
weather_bike_df = hour_bike_df.groupby(by='weathersit').agg({
    "instant" : "nunique",
    "cnt" : "sum"
})
weather_bike_df.columns = ['cnthour', 'cntpeople']
weather_bike_df['Mean'] = weather_bike_df['cntpeople'] / weather_bike_df['cnthour']
weather_bike_df['weathersit'] = ['Clear', 'Cloudy','Light', 'Heavy']
weather_bike_df
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Cuaca yang cocok bagi para pesepeda?

a. Banyaknya waktu yang terjadi pada cuaca tertentu

```python
fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(30,6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="cnthour", x="weathersit", data=weather_bike_df, palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Hours", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

sns.barplot(y="cntpeople", x="weathersit", data=weather_bike_df, palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By People", loc="center", fontsize=18)
ax[1].tick_params(axis ='x', labelsize=15)

sns.barplot(y="Mean", x="weathersit", data=weather_bike_df, palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Average", loc="center", fontsize=18)
ax[2].tick_params(axis ='x', labelsize=15)

plt.suptitle("Best Weather for Cycling", fontsize=20)
plt.show()
```

### Pertanyaan 2: Waktu yang paling sering untuk bersepeda?

```python
# Mencari season shift
hour_bike_df['season_shift'] = hour_bike_df['season'].shift(1)

# Temukan baris di mana season berubah (yaitu, ketika season tidak sama dengan season_shift)
season_changes = hour_bike_df[hour_bike_df['season'] != hour_bike_df['season_shift']]

# Hanya ambil kolom dteday dan season
season_changes = season_changes[['dteday', 'season']]
season_changes['dteday'] = season_changes['dteday'].dt.strftime('%Y-%m')
season_changes.rename(columns={
    'dteday' : 'dtemonth'
},inplace=True)

season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Autumn', 4: 'Winter'}
season_changes['season'] = season_changes['season'].map(season_mapping)

# Lihat hasilnya
print(season_changes)
```

```python
plt.figure(figsize=(20,8))
plt.plot(monthly_bike_df['dtemonth'], monthly_bike_df['cnt'])

colors = ['green', 'orange', 'red', 'blue', 'green', 'orange', 'red', 'blue', 'green']

for color,season,season_change in zip(colors,season_changes['season'],season_changes['dtemonth']):
    plt.axvline(x=season_change, color=color, linestyle='--', label=season)
    plt.text(season_change, 150000, season, fontsize=15, color=color, ha='center')

plt.title("Number of Cycler per Month (2011-2012)", loc='center', fontsize=25)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
